```python
import numpy as np
from time2npy import tiq2npy
from pylab import specgram, cm, psd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dic = tiq2npy('signal.tiq', 1000)
x = dic['data']
fs = dic['fs']
lframes = dic['lframes']
center = dic['center']
```

```python
def freqPlot(x, fs, c):
    n = x.size
    ts = 1.0/fs
    f = np.fft.fftfreq(n, ts) + c
    y = np.fft.fft(x)/n
    plt.plot(f, 10*np.log10(pow(abs(y),2)))
```

```python
freqPlot(x, fs, center)
```

```python
Pxx, freqs = psd(x, NFFT=lframes, Fs = fs, noverlap=512)
```

```python
Pxx, freqs, bins, im = specgram(x, NFFT=lframes, Fs=fs, noverlap=512, cmap=cm.jet)
```
