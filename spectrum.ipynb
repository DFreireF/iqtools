# Spectrum calculation of Schottky signals

```python
import numpy as np
from time2npy import tiq2npy
from pylab import specgram, cm, psd
import matplotlib.pyplot as plt
from cmath import phase
%matplotlib inline
```

```python
def freqPlot(x, fs, c):
    n = x.size
    ts = 1.0/fs
    f = np.fft.fftfreq(n, ts) + c
    y = np.fft.fft(x)/n
    plt.plot(f, 10*np.log10(pow(abs(y),2)), '.')
    plt.xlabel("Frequency [Hz]")
    plt.ylabel("Power Spectral Density [dB/Hz]")
    plt.grid(True)
    return f, y
```

```python
dic1 = tiq2npy('/Users/shahab/Desktop/RSA51_238U90+190MeVu-2014.10.23.14.58.52.007.TIQ', 200)
x1 = dic1['data']
fs1 = dic1['fs']
lframes1 = dic1['lframes']
center1 = dic1['center']
```

```python
dic2 = tiq2npy('/Users/shahab/Desktop/RSA52_238U90+190MeVu_72uA-2014.10.23.14.59.07.685.TIQ', 200)
x2 = dic2['data']
fs2 = dic2['fs']
lframes2 = dic2['lframes']
center2 = dic2['center']
```

```python
f1, y1 = freqPlot(x1, fs1, 0)
f2, y2 = freqPlot(x2, fs2, 0)
```

```python
Pxx1, freqs1 = psd(x1, NFFT=lframes1, Fs = fs1, noverlap=512)
Pxx2, freqs2 = psd(x2, NFFT=lframes2, Fs = fs2, noverlap=512)
```
