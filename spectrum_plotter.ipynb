# Plot of spectra

```python
import numpy as np
from pylab import psd
import matplotlib.pyplot as plt
%matplotlib inline
from iqtools import *
```

```python
filename = '/Users/shahab/test/p400332/acq_data.tiq'
dic1,_ = read_tiq(filename, 100, 1024, 1)
center1 = dic1['center']
fs1 = dic1['fs']
nframes_tot = dic1['nframes_tot']
x1 = dic1['data']
rbw1 = dic1['rbw']
```

```python
data_csv = read_data_csv('/Users/shahab/test/p400332/acq_data_export.csv')
```

```python
f1, v1, p1 = get_fft_50(x1, fs1, 0)
f_csv, v_csv, p_csv = get_fft_50(data_csv, fs1, 0)
f_res, p_res = read_result_csv('/Users/shahab/test/p400332/result_export_rbw_1.csv')
```

```python
plt.plot(f1, 10*np.log10(p1*1000),  label = 'Data TIQ')
plt.plot(f_csv, 10*np.log10(p_csv*1000), label = 'Data CSV')
plt.plot(f_res,p_res, label = 'Result CSV RBW = 1Hz')
plt.grid(True)
plt.title('Both using 50 ohm termination')
plt.xlabel('Frequency [Hz]')
plt.ylabel('PSD [dBm/Hz]')
plt.legend(loc=9, bbox_to_anchor=(0.5, -0.1))
#plt.savefig('/Users/shahab/Desktop/both50ohm.pdf')
```

```python
f1, v1, p1 = get_fft_1(x1, fs1, 0)
f_csv, v_csv, p_csv = get_fft_1(data_csv, fs1, 0)
f_res, p_res = read_result_csv('/Users/shahab/test/p400332/result_export_rbw_1.csv')
```

```python
plt.plot(f1, 10*np.log10(p1*1000),  label = 'Data TIQ')
plt.plot(f_csv, 10*np.log10(p_csv*1000), label = 'Data CSV')
plt.plot(f_res,p_res, label = 'Result CSV RBW = 1Hz')
plt.grid(True)
plt.title('Both using 1 ohm termination')
plt.xlabel('Frequency [Hz]')
plt.ylabel('PSD [dBm/Hz]')
plt.legend(loc=9, bbox_to_anchor=(0.5, -0.1))
#plt.savefig('/Users/shahab/Desktop/both1ohm.pdf')
```
