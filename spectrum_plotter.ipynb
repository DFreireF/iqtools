```python
from iqtools import *
import numpy as np
from pylab import psd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename = '/Users/shahab/test/p_400_MeV_u_332_uA_0.tiq'
dic1,_ = read_tiq(filename, 100, 1024, 1)
center1 = dic1['center']
fs1 = dic1['fs']
nframes_tot = dic1['nframes_tot']
x1 = dic1['data']
```

```python
f, v, p = plot_fft(x1, fs1, 0)
channel_power_dbm(f, p)
```

```python
fwelch, pwelch = plot_pwelch(x1, fs = fs1)
#plt.xlim(-30000, 50000)
#plt.ylim(-106, -60)
plt.title(filename)
plt.xlabel('Rel. Freq [Hz] @ {} [MHz] --> 124th harmonic'.format(round(center1/1e6, 2)))
plt.savefig(filename+'.pdf')
channel_power_dbm(fwelch, pwelch)
```
