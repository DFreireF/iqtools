```python
from iqtools import *
%matplotlib inline
```

```python
filename = 'p_400_MeV_u_332_uA_0.tiq'
```

```python
iq = get_iq_object(filename)
iq.read_samples(200*1024)
xx, yy, zz = iq.get_spectrogram(lframes=1024, nframes=200)
iq.method='mtm'
plot_spectrogram(xx, yy, zz, cen=iq.center, title=iq.file_basename)
```
